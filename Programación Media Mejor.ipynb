## Operador Indice []

Como siempre se ha hablado de "cadena de caracteres", se puede accesar a cada eslabón (elemento) de dicha cadena por medio de su operador de índice utilizando la sintaxis:

s[index]

Dichos índices tienen su base en 0, esta convención es diferente a la utilizada por software como Matlab y permite una interacción limpia con la notación matemática.

Los índices son extremadamente importantes: pueden utilizarse de muchas maneras, como por ejemplo recuperar un elemento. Muchos cálculos implican procesar una cadena carácter por carácter. A menudo empiezan por el principio, seleccionan cada carácter por turno, hacen algo con él y siguen hasta el final. 

Este patrón de proceso se llama recorrido.

```python
# Ejm.10
s = "Programming"
for i in range(0, len(s), 3):
    print s[i],
```

Empleando la función for para i en una lista que admite tres argumentos se manipula el objeto a conveniencia.

El primer  argumento indica la posicion del elemento desde donde se quiere comenzar a manipular el objeto, el segundo argumento indica la longitud de la cadena, cuantos carácteres se tendrán en cuenta en la modificación y el tercer argumento indica cada cuantas posiciones se tomará un carácter para obtener la cadena final(intervalo).

Los indices pueden darse en cualquier dirección! El paso puede darse con valores positivos y negativos.

```python
# Ejm.11
s = "Programming"
print s[-1]
print s[len(s)-1]
print s[-len(s)]
```

Como puede observase en s [-1] el conteo de derecha a isquierda en la cadena no tiene su base en cero, sino en 1, es decir, en -1. Con [len(s)-1], len(s) recorre la cadena de isquierda a derecha y toma el primer cáracter al devolverse. Así, con [-len(s)] se recorre la cadena de derecha a isquierda y toma el ultimo cáracter.

```python
# Ejm.12
s = "Programming"
for i in range(-1,-len(s)-1,-3):
    print s[i],

# ---------------------------------------------------------------
# Así como en el ejemplo 11, tambien se pueden implementar valores
# negativos al ejemplo 10.
# ---------------------------------------------------------------
```

# El operador rebanada: **[Inicio : Final-1]**

A la subcadena de una cadena se le llama porción. La selección de una porción es similar a la selección un carácter. El operador rebanada regresa un pedazo de la cadena de caracteres utilizando la sintaxis s[Inicio : Final-1]. El operador [n:m] devuelve la parte de la cadena desde el “n-ésimo” carácter hasta el “m-ésimo”, incluyendo el primero pero excluyendo el último.

```python
# Ejm.13
s = "Programming"
print s[0:len(s)]
print s[0:]
print s[:]  
print s[4:8]
print s[2:-2]
```

Si se omite el primer índice (antes de los dos puntos), la porción comienza al principio de la cadena. Si omite el segundo índice, la porción llega al final de la cadena. Si se omiten ambos, toma los valores base 0 y -1.

Tambien es posible tomar rebanadas excluyendo cáracteres a cada lado de la cadena. Véase s[2:-2]; recorre de isquierda a derecha en el valor incial y recorre de derecha a isquierda en el valor final.

# Buscar sub-cadenas

Se puede buscar una sub-cadena (un pedazo) dentro de un cadena utilzando:

| str | Regresa...
|----|----
|endswith(s1: str): bool | Verdadero si la cadena termina con la sub-cadena s1.
|startswhith(s1: str): bool | Verdadero si la cadena empieza con la sub-cadena s1.
|find(s1): int | El índice más bajo donde s1 comienza en la cadena o -1 si no se encuentra en la cadena.
|rfind(s1): int | El índice más alto donde s1 comienza en la cadena o -1 si no se encuentra en la cadena.
|count(s1): int | Regresa el número de ocurrencias de la sub-cadena (sin solaparse).

```python
# Ejm.29
s = "Bienvenido a Python"
print s.endswith("thon")
print s.startswith("bien")
print s.find("veni")
print s.find("pyton")

# ---------------------------------------------------------------
# Implementación con métodos de búsqueda. Las respuestas a las
# relaciónes logícas son de tipo boooleno o de tipo entero.
# ---------------------------------------------------------------
```

# Convertiendo Cadenas

Se puede hacer una copia de una cadena utilizando métodos que permiten controlar la “capitalización” de las letras en una copia de la cadena de caracteres o reemplazar la cadena por otra cosa.


| str | Regresa...
|----|----
|capitalize(): str | Regresa una copia de la cadena con sólo el primer carácter en mayúscula.
