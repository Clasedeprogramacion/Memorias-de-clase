## Escribiendo un programa simple

Nos plantemos el problema de computar el radio de un círculo.
Escribir un programa involucra diseñar un algoritmo y luego traducirlo a instrucciones o código. Cuando se *programa (code)* es cuando se escribe un programa. Un algoritmo describe como el problema es solucionado listando las acciones necesarias que se necesitan para que el software se ejecute. 
Los algoritmos ayudan al programador antes de escribir un programa.
Los algoritmos se escriben en código natural o *pseudo-código (lenguaje natural mezclado con algo de código de programación)*.
* El algoritmo para un código de calcular el área del circulo podría ser el siguiente:
    - Obtenga el radio del circulo del usuario.
    - Compute el área aplicando la fórmula $Area=radio \times radio \times \pi$.
    - Muestre el resultado del cálculo

Como el programa necesita leer el radio, se necesita que el usuario lo introduzca por medio del teclado, para esto se necesitan dos cosas:
* Leer el radio.
* Almacenar el valor del radio en el programa.
Es decir, el valor del radio se almacena en la memoria del computador, para accesarlo el programa necesita utilizar una *variable*. Una variable es el nombre que referencia al valor almacenado en la memoria. 
En lugar de utilizar nombres como **x** o **y** es recomendado utilizar nombres descriptivos.

```python
"""
Este programa recibe el valor del radio del usuario y 
regresa el valor del área presentado en la terminal
por medio de una cadena de caracteres.
"""

# Asignar un valor al radio.
radio = 20 # la variable radio ahora tiene un valor igual 20.

# Computar el area de un circulo.
area = radio * radio * 3.14159
# Otra forma de computarlo es mediante: area = radio ** 2 * 3.14159

"""
tenemos la siguiente operación: int * int * float, 
por tanto la variable area se promueve a float o flotante
y no se comete un error lógico.
"""
"""
La función Print muestra reultados como: cadena de caracteres, igualdades de variables,
operaciones matemáticas y demas; sin embargo, lo hace con su debida sintaxis para cada
una y separadas por comas en caso de que hayan diferentes tipos de resultados que se 
quieran mostrar.
"""
# Muestra los resultados.
print ("El área del circulo de radio ", radio, "es", area)
```

La función print también permite realizar operaciones matemáticas dentro de ella misma, por ejemplo:


```python
radio = 20
print ("El área del circulo de radio ", radio, "es", radio * radio * 3.14159)
```

# Identificadores

Los identificadores son nombres que identifican variables y funciones en un programa, por ejemplo: area, radio, print, input. 
Obedecen las siguientes reglas:
* Consisten sólamente de caracteres que contienen:
    * Letras
    * Números
    * El caracter sub-indice "_".
* Un identificador TIENE que empezar con una letra o un sub-indice, no un número.
* Un identificador no puede ser una palabra reservada (import, print, eval, input, etc)
* Un identificador puede ser tan largo como se quiera o se necesite.
    

Area, area y AREA son 3 identificadores diferentes.

# Asignaciones simultaneas

En los lenguajes de programación se puede usar un identificador intermedio entre las declaraciones de asignación para hacer un intercambio de variable

```python
x = 1
y = 2
print (x, y)

temp = x
x = y
y = temp
print (x, y)
```

Sin embargo Python permite realizar una organización en un solo estatuto que permita este mismo intercambio de variables de manera sencilla y elegante.
A esto se le denomina **Asignación Simultanea**

```python
x = 1
y = 2
print x, y

x, y = y, x
print x, y
```

# Tipos de datos numéricos y operadores

Los tipos predeterminados de datos son enteros y números de tipo flotante y se pueden trabajar con los operadores:
* "**" Exponenciación
* "*" Multiplicación
* "/" División flotante
* "//" División entera
* "%" Resto
* "+" Adicion
* "-" Sustracción

Estos operadores estan organizados de forma jerarquica, de manera que el programa al leer el código de izquierda a derecha, comienza operando exponenciales, luego multiplicaciones y asi sucesivamente hasta llear a sustracciones.

*Ademas de estos operadores, existe un caracter que encabeza la jerarquia (los parentesis) debido a que tambien se utilizan para organizar o dar prioridad a algunos elementos de la expresión

###### Tipos de datos numericos
Los predeterminados son enteros y flotantes, sin embargo puedo intercambiar entre ellos usando la función int() que convierte todo numero a entero y la función float() que convirte todo numero a flotante. Por otro lado existe la función str() que convierte todos los datos en una cadena de caracteres o string. otros tipos de datos numericos son los complejos y vienen dados de la forma "a = 4 + 5j"

## Cual es el resultado de las siguietes operaciones?

```python
42 / 5
42 // 5
42 % 5
40 % 5
1 % 2
2 % 1
45 + 4 * 4 - 2
45 + 43 % 5 * (23 * 3 % 2)
5 ** 2
5.1 ** 2
254.**10000
```

```python
print ("a.", 42 / 5)
print ("b.", 42 // 5)
print ("c.", 42 % 5)
print ("d.", 40 % 5)
print ("e.", 1 % 2)
print ("f.", 2 % 1)
print ("g.", 45 + 4 * 4 - 2)
print ("h.", 45 + 43 % 5 * (23 * 3 % 2))
print ("i.", 5 ** 2)
print ("j.", 5.1 ** 2)
print ("k.", 254.**10000) #Esta operación da un flotante con una cantidad de digitos decimales muy larga para poder ser expresada
```
